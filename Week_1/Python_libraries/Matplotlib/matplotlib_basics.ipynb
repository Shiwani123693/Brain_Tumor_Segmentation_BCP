```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot multiple lines
plt.plot(x, y1, label="y = x^2", color="blue", linestyle="--", marker="o")
plt.plot(x, y2, label="y = x", color="red", linestyle="-.", marker="s")

# Add labels, legend, and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

# Show the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

# Create scatter plot
plt.scatter(x, y, color="green", marker="x", s=100)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")

# Show the plot
plt.show()
```

```python
# Data
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

# Create bar chart
plt.bar(categories, values, color="purple")

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")

# Show the plot
plt.show()
```

```python
# Create horizontal bar chart
plt.barh(categories, values, color="orange")

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Bar Chart")

# Show the plot
plt.show()
```

```python
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]

# Create histogram
plt.hist(data, bins=5, color="blue", edgecolor="black")

# Add labels and title
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")

# Show the plot
plt.show()
```

```python
# Data
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 30, 20, 10]
colors = ["gold", "lightblue", "lightgreen", "pink"]
explode = (0.1, 0, 0, 0)  # Highlight the first slice

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y, color="red", linewidth=2, linestyle="--", marker="o", markersize=8)

# Add grid, labels, and title
plt.grid(True, linestyle="--", alpha=0.5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Plot")

# Show the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Create subplots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y1, color="blue")
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.plot(x, y2, color="green")
plt.title("Plot 2")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Assignment 1
'''
Create a numpy array of length 1000 which resembles a normal distribution with mean 0 and standard deviation 1.
Create 2 subplots both histograms with 20 bins side by side, one which has a beta (a=2, b=5) distribution, the other which has the normal distribution.
'''
```

```python
# TODO
import numpy as np
import matplotlib.pyplot as plt
normal_data = np.random.normal(loc=0, scale=1, size=1000)
beta_data = np.random.beta(a=2, b=5, size=1000)
plt.figure(figsize=(10, 4))
# Beta distribution histogram
plt.subplot(1, 2, 1)
plt.hist(beta_data, bins=20)
plt.title("Beta Distribution (a=2, b=5)")

# Normal distribution histogram
plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=20)
plt.title("Normal Distribution (mean=0, std=1)")

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
# Assignment 2
'''
Generate a NumPy array of 1000 random values drawn from a normal distribution with mean 0 and std. dev 1
Plot a histogram of the data with 20 bins.
Add vertical lines to indicate the mean of the data, the 25th percentile and the 75th percentile.
Annotate the plot with labels for the mean, 25th percentile, and 75th percentile.
Customize the visualization (light pink colour scheme, titles, labels, legend).
'''
```

```python
# TODO

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Calculate statistics
mean_val = np.mean(data)
p25 = np.percentile(data, 25)
p75 = np.percentile(data, 75)

# Step 3: Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='lightpink', edgecolor='black', alpha=0.7)

# Step 4: Add vertical lines
plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
plt.axvline(p25, color='blue', linestyle='--', label=f'25th Percentile: {p25:.2f}')
plt.axvline(p75, color='green', linestyle='--', label=f'75th Percentile: {p75:.2f}')

# Step 5: Add annotations
plt.text(mean_val, plt.ylim()[1]*0.9, 'Mean', color='red', ha='center')
plt.text(p25, plt.ylim()[1]*0.8, '25th Percentile', color='blue', ha='center')
plt.text(p75, plt.ylim()[1]*0.8, '75th Percentile', color='green', ha='center')

# Step 6: Customize plot
plt.title('Histogram of Normally Distributed Data', fontsize=16)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)

# Step 7: Show plot
plt.show()
```
