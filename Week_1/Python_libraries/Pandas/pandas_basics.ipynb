```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the CSV
df = pd.read_csv('videogamesales.csv')

# Step 2: Add a 'global_sales' column (sum of all regions)
df['Global_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']

# Step 3: Sort the DataFrame by global sales (highest first)
df_sorted = df.sort_values(by='Global_Sales', ascending=False)
print("DataFrame sorted by Global Sales:\n", df_sorted.head(10))  # printing top 10 for readability

# Step 4: Plot total number of copies sold of each genre globally
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title("Total Global Sales by Genre")
plt.ylabel("Global Sales (millions)")
plt.xlabel("Genre")
plt.xticks(rotation=45)
plt.show()

# Step 5: Filter games containing 'Grand Theft Auto'
gta_games = df[df['Name'].str.contains('Grand Theft Auto', case=False, na=False)]
gta_df = gta_games[['Name', 'Platform', 'Year', 'EU_Sales', 'JP_Sales']].copy()
gta_df['EU+JP_Sales'] = gta_df['EU_Sales'] + gta_df['JP_Sales']
gta_df = gta_df[['Name', 'Platform', 'Year', 'EU+JP_Sales']]
print("\nGrand Theft Auto games (EU + JP sales):\n", gta_df)

# Step 6: Pie chart of total sales of all GTA games combined by region
total_gta_sales = gta_games[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
total_gta_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7), startangle=90)
plt.title("Total Sales of Grand Theft Auto Games by Region")
plt.ylabel("")
plt.show()
```
